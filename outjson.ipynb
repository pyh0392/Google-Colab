```python
import json

input_file = "sunwukong_dialogues_concurrent.jsonl"  # DeepSeek æå–çš„åŸå§‹ JSONL
output_file = "sunwukong_only.jsonl"

train_data = []

# æŒ‰ chunk_id åˆ†ç»„
from collections import defaultdict
chunks = defaultdict(list)
with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        chunks[data["chunk_id"]].append(data)

# æ„å»º instruction -> output
for chunk_id, dialogues in chunks.items():
    # ä¿è¯å¯¹è¯é¡ºåº
    dialogues.sort(key=lambda x: x["dialogue_index"])
    for i in range(len(dialogues)-1):
        # output å¿…é¡»æ˜¯å­™æ‚Ÿç©º
        if dialogues[i+1]["role"] in ["çŸ³çŒ´", "å­™æ‚Ÿç©º", "æ‚Ÿç©º"]:
            instruction = f"[{dialogues[i]['role']}]: {dialogues[i]['dialogue']}"  # ä¸Šä¸€å¥ä»»æ„è§’è‰²
            output = f"[{dialogues[i+1]['role']}]: {dialogues[i+1]['dialogue']}"   # ä¸‹ä¸€å¥å­™æ‚Ÿç©º
            train_data.append({
                "instruction": instruction,
                "input": "",
                "output": output
            })

# ä¿å­˜ä¸º JSONL
with open(output_file, "w", encoding="utf-8") as f:
    for item in train_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(f"âœ… å·²ç”Ÿæˆ {len(train_data)} æ¡è®­ç»ƒæ ·æœ¬ï¼Œä¿å­˜è‡³ {output_file}")
```
