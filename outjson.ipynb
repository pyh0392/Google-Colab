```python
import json

input_file = "sunwukong_dialogues_concurrent.jsonl"  # DeepSeek 提取的原始 JSONL
output_file = "sunwukong_only.jsonl"

train_data = []

# 按 chunk_id 分组
from collections import defaultdict
chunks = defaultdict(list)
with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        chunks[data["chunk_id"]].append(data)

# 构建 instruction -> output
for chunk_id, dialogues in chunks.items():
    # 保证对话顺序
    dialogues.sort(key=lambda x: x["dialogue_index"])
    for i in range(len(dialogues)-1):
        # output 必须是孙悟空
        if dialogues[i+1]["role"] in ["石猴", "孙悟空", "悟空"]:
            instruction = f"[{dialogues[i]['role']}]: {dialogues[i]['dialogue']}"  # 上一句任意角色
            output = f"[{dialogues[i+1]['role']}]: {dialogues[i+1]['dialogue']}"   # 下一句孙悟空
            train_data.append({
                "instruction": instruction,
                "input": "",
                "output": output
            })

# 保存为 JSONL
with open(output_file, "w", encoding="utf-8") as f:
    for item in train_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(f"✅ 已生成 {len(train_data)} 条训练样本，保存至 {output_file}")
```
